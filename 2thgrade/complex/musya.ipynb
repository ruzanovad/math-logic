```python
import numpy as np
import matplotlib.pyplot as plt

# Define the Schwarz-Christoffel transformation
def schwarz_christoffel(z):
    return (1 - np.exp(np.pi * 1j * z)) / (1 + np.exp(np.pi * 1j * z))

# Define the region to plot
n = 1000
x = np.linspace(-2, 3, n)
y = np.linspace(-2, 2, n)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Find the points inside the region
inside = (np.abs(np.abs(Z) - 1) < 1) & (np.abs(Z - 1j) < 1)

# Apply the transformation to the points inside the region
Z_inside = schwarz_christoffel((Z[inside] - 1j) / 2)

# Plot the transformed region
plt.figure(figsize=(8, 8))
plt.plot(Z_inside.real, Z_inside.imag, '.', markersize=1)
plt.axis('equal')
plt.axis('off')
plt.show()
```

```python

import numpy as np
import matplotlib.pyplot as plt

# Создание массива точек на комплексной плоскости
x = np.linspace(-1, 1, 1000)
y = np.linspace(-1, 1, 1000)
X, Y = np.meshgrid(x, y)
Z = X + 1j * Y

# Отображение точек через w = sqrt(z)
W = Z

# Разделение на верхнюю и нижнюю полуплоскости
upper_half_plane = np.log(np.where(W.imag > 0, W, np.nan))
lower_half_plane = np.log(np.where(W.imag <= 0, W, np.nan))

# Построение графика
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111)

# Построение осей
ax.axhline(0, color='black', linewidth=0.5)
ax.axvline(0, color='black', linewidth=0.5)

t = np.linspace(0.0001, 1)
first_cut = t
second_cut = np.cos(2*np.pi/3)*t + 1j*np.sin(2*np.pi/3)*t
third_cut = np.cos(-2*np.pi/3)*t + 1j*np.sin(-2*np.pi/3)*t

T = np.linspace(0.00001, 2*np.pi/6)

fc = np.cos(T-np.pi/6) + 1j * np.sin(T-np.pi/6)
sc = np.cos(T+np.pi/2) + 1j * np.sin(T+np.pi/2)
tc = np.cos(T+np.pi*7/6) +1j* np.sin(T+np.pi*7/6)



def func(X):
    # return np.power(X, 2)
    return np.log(X)
first_cut = func(first_cut)
second_cut = func(second_cut)
third_cut = func(third_cut)

fc = func(fc)
sc = func(sc)
tc = func(tc)


# Построение разреза по радиусу [0,1]
ax.plot(first_cut.real, first_cut.imag, color='red', alpha=0.7, linestyle='-', linewidth=3)
ax.plot(second_cut.real, second_cut.imag, color='purple', alpha=0.7, linestyle='-', linewidth=3)
ax.plot(third_cut.real, third_cut.imag, color='yellow', alpha=0.7, linestyle='-', linewidth=5)
# ax.plot([0, 1], [0, 0], color='red', alpha=0.7, linestyle='-', linewidth=3)

ax.plot(fc.real, fc.imag, color='red', alpha=0.7, linestyle='-', linewidth=3)
ax.plot(sc.real, sc.imag, color='purple', alpha=0.7, linestyle='-', linewidth=3)
ax.plot(tc.real, tc.imag, color='yellow', alpha=0.7, linestyle='-', linewidth=5)

# Построение верхней полуплоскости
ax.scatter(upper_half_plane.real, upper_half_plane.imag, color='blue', s=1, alpha=0.5)

# Построение нижней полуплоскости
ax.scatter(lower_half_plane.real, lower_half_plane.imag, color='gray', s=1, alpha=0.5)

# Настройка пределов осей
ax.set_xlim(-1.5, 1.5)
ax.set_ylim(-1.5, 1.5)

# Отображение
plt.show()
```
