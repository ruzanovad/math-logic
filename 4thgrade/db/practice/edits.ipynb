# правки в план
## факультет
- убрать из списка атрибутов код факультета 
- заменить атрибут **декан** на **текущий декан**
- заменить атрибут **контактные данные** на **контактные данные деканата**
- учесть изменения факультета по годам (в разные годы разные учебные планы, деканы, реализуемые программы, и т.д.)

**idea**: может быть ввести сущность `деканат` и ссылаться на эту таблицу из таблицы `факультет`?

## дисциплина
- добавить атрибут **тип дисциплины**: обязательная, предмет по выбору (мб какая-о ещё)
- часы, отведённые на изучение дисциплины
- форма контроля знаний (зачёт, экз, дифзач)

## документ
- добавить атрибут **дата назначения документа**


## general recommendation
- хранить списки чего-то в форме таблицы с id, а не в форме полей (потому что нарушается 1 НФ)
- убрать списки дисциплин (хранить учебный план)
- нужно хранить мало информации о специальности, но много об учебном плане. Учесть ежегодное изменение учплана
- студенты могут учиться в нескольких группах (на нескольких направлениях и даже факультетах)
- добыить атрибут **номер зачетки** студента (непонятно как, учитывая предыдущее замечание)
- что делать с неучебными подразделениями
- сотрудники могут занимать несколько должностей на рахных кафедрах и факультетах
- библиотека должна знать об учебных планах, чтобы корректно выдавать литературу
